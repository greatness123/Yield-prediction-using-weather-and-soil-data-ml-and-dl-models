```python
"""
1D Convolutional Neural Network for Crop Yield Prediction
Using Weather and Soil Data from Indian Agriculture Dataset (1997-2020)

This script implements a 1D CNN model to predict crop yield based on:
- Weather data: Annual rainfall
- Soil data: Fertilizer and Pesticide usage (proxies for soil composition)
- Agricultural inputs: Area, Crop type, Season, State
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.utils import to_categorical
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print("=" * 80)
print("1D CNN MODEL FOR CROP YIELD PREDICTION")
print("=" * 80)

# ============================================================================
# 1. DATA LOADING AND EXPLORATION
# ============================================================================
print("\n[1] Loading and Exploring Data...")

# Load dataset
df = pd.read_csv('/content/crop_yield.csv')

print(f"Dataset shape: {df.shape}")
print(f"\nColumns: {list(df.columns)}")
print(f"\nBasic statistics:")
print(df.describe())

# Check for missing values
print(f"\nMissing values:")
print(df.isnull().sum())

# ============================================================================
# 2. DATA PREPROCESSING
# ============================================================================
print("\n[2] Preprocessing Data...")

# Remove any rows with missing values (if any)
df = df.dropna()

# Create a copy for processing
df_processed = df.copy()

# Encode categorical variables
label_encoders = {}
categorical_cols = ['Crop', 'Season', 'State']

for col in categorical_cols:
    le = LabelEncoder()
    df_processed[col + '_encoded'] = le.fit_transform(df_processed[col])
    label_encoders[col] = le
    print(f"Encoded {col}: {len(le.classes_)} unique values")

# ============================================================================
# 3. FEATURE ENGINEERING
# ============================================================================
print("\n[3] Feature Engineering...")

# Select features for the model
# Weather: Annual_Rainfall
# Soil/Agricultural inputs: Fertilizer, Pesticide, Area
# Categorical: Crop, Season, State (encoded)
# Temporal: Crop_Year

feature_columns = [
    'Crop_Year',
    'Area',
    'Annual_Rainfall',
    'Fertilizer',
    'Pesticide',
    'Crop_encoded',
    'Season_encoded',
    'State_encoded'
]

# Create feature matrix
X = df_processed[feature_columns].values
y = df_processed['Yield'].values

print(f"Feature matrix shape: {X.shape}")
print(f"Target vector shape: {y.shape}")

# Handle outliers in yield (remove extreme outliers)
q1 = np.percentile(y, 25)
q3 = np.percentile(y, 75)
iqr = q3 - q1
lower_bound = q1 - 3 * iqr
upper_bound = q3 + 3 * iqr

outlier_mask = (y >= lower_bound) & (y <= upper_bound)
X = X[outlier_mask]
y = y[outlier_mask]

print(f"After removing extreme outliers: {X.shape[0]} samples")

# ============================================================================
# 4. TRAIN-TEST SPLIT
# ============================================================================
print("\n[4] Splitting Data...")

# Split into train, validation, and test sets
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42, shuffle=True
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.176, random_state=42, shuffle=True
)  # 0.176 * 0.85 ≈ 0.15 of total

print(f"Training set: {X_train.shape}")
print(f"Validation set: {X_val.shape}")
print(f"Test set: {X_test.shape}")

# ============================================================================
# 5. FEATURE SCALING
# ============================================================================
print("\n[5] Scaling Features...")

# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# Reshape for 1D CNN (samples, time_steps, features)
# For 1D CNN, we treat each feature as a time step
X_train_cnn = X_train_scaled.reshape(X_train_scaled.shape[0], X_train_scaled.shape[1], 1)
X_val_cnn = X_val_scaled.reshape(X_val_scaled.shape[0], X_val_scaled.shape[1], 1)
X_test_cnn = X_test_scaled.reshape(X_test_scaled.shape[0], X_test_scaled.shape[1], 1)

print(f"Reshaped data for 1D CNN:")
print(f"  Training: {X_train_cnn.shape}")
print(f"  Validation: {X_val_cnn.shape}")
print(f"  Test: {X_test_cnn.shape}")

# ============================================================================
# 6. BUILD 1D CNN MODEL
# ============================================================================
print("\n[6] Building 1D CNN Model...")

def build_1d_cnn_model(input_shape):
    """
    Build a 1D Convolutional Neural Network for yield prediction

    Architecture:
    - Conv1D layers to extract local patterns from features
    - MaxPooling for dimensionality reduction
    - Dropout for regularization
    - Dense layers for final prediction
    """
    model = models.Sequential([
        # First Convolutional Block
        layers.Conv1D(filters=64, kernel_size=3, activation='relu',
                     padding='same', input_shape=input_shape),
        layers.BatchNormalization(),
        layers.MaxPooling1D(pool_size=2),
        layers.Dropout(0.3),

        # Second Convolutional Block
        layers.Conv1D(filters=128, kernel_size=3, activation='relu',
                     padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling1D(pool_size=2),
        layers.Dropout(0.3),

        # Third Convolutional Block
        layers.Conv1D(filters=256, kernel_size=2, activation='relu',
                     padding='same'),
        layers.BatchNormalization(),
        layers.GlobalMaxPooling1D(),
        layers.Dropout(0.4),

        # Fully Connected Layers
        layers.Dense(128, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.4),

        layers.Dense(64, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.3),

        layers.Dense(32, activation='relu'),
        layers.Dropout(0.2),

        # Output Layer
        layers.Dense(1, activation='linear')
    ])

    return model

# Build the model
input_shape = (X_train_cnn.shape[1], 1)
model = build_1d_cnn_model(input_shape)

# Compile the model
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='huber',  # Huber loss is robust to outliers
    metrics=['mae', 'mse']
)

# Display model architecture
print("\nModel Architecture:")
model.summary()

# ============================================================================
# 7. TRAIN THE MODEL
# ============================================================================
print("\n[7] Training the Model...")

# Define callbacks
early_stopping = callbacks.EarlyStopping(
    monitor='val_loss',
    patience=20,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=10,
    min_lr=1e-7,
    verbose=1
)

# Train the model
history = model.fit(
    X_train_cnn, y_train,
    validation_data=(X_val_cnn, y_val),
    epochs=150,
    batch_size=64,
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

# ============================================================================
# 8. EVALUATE THE MODEL
# ============================================================================
print("\n[8] Evaluating the Model...")

# Make predictions
y_train_pred = model.predict(X_train_cnn).flatten()
y_val_pred = model.predict(X_val_cnn).flatten()
y_test_pred = model.predict(X_test_cnn).flatten()

# Calculate metrics
def calculate_metrics(y_true, y_pred, dataset_name):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    print(f"\n{dataset_name} Metrics:")
    print(f"  RMSE: {rmse:.4f}")
    print(f"  MAE:  {mae:.4f}")
    print(f"  R²:   {r2:.4f}")

    return {'RMSE': rmse, 'MAE': mae, 'R2': r2}

train_metrics = calculate_metrics(y_train, y_train_pred, "Training")
val_metrics = calculate_metrics(y_val, y_val_pred, "Validation")
test_metrics = calculate_metrics(y_test, y_test_pred, "Test")

# ============================================================================
# 9. VISUALIZATION
# ============================================================================
print("\n[9] Creating Visualizations...")

# Set style
plt.style.use('default')
sns.set_palette("husl")

# Create figure with subplots
fig = plt.figure(figsize=(20, 12))

# 1. Training History - Loss
ax1 = plt.subplot(3, 3, 1)
plt.plot(history.history['loss'], label='Training Loss', linewidth=2)
plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
plt.title('Model Loss Over Epochs', fontsize=14, fontweight='bold')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

# 2. Training History - MAE
ax2 = plt.subplot(3, 3, 2)
plt.plot(history.history['mae'], label='Training MAE', linewidth=2)
plt.plot(history.history['val_mae'], label='Validation MAE', linewidth=2)
plt.title('Mean Absolute Error Over Epochs', fontsize=14, fontweight='bold')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('MAE', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

# 3. Predictions vs Actual (Test Set)
ax3 = plt.subplot(3, 3, 3)
plt.scatter(y_test, y_test_pred, alpha=0.5, s=20)
min_val = min(y_test.min(), y_test_pred.min())
max_val = max(y_test.max(), y_test_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2, label='Perfect Prediction')
plt.title(f'Test Set: Predicted vs Actual\nR² = {test_metrics["R2"]:.4f}',
          fontsize=14, fontweight='bold')
plt.xlabel('Actual Yield', fontsize=12)
plt.ylabel('Predicted Yield', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

# 4. Residuals Distribution (Test Set)
ax4 = plt.subplot(3, 3, 4)
residuals = y_test - y_test_pred
plt.hist(residuals, bins=50, edgecolor='black', alpha=0.7)
plt.title('Residuals Distribution (Test Set)', fontsize=14, fontweight='bold')
plt.xlabel('Residual (Actual - Predicted)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.axvline(x=0, color='r', linestyle='--', linewidth=2)
plt.grid(True, alpha=0.3, axis='y')

# 5. Residuals vs Predicted (Test Set)
ax5 = plt.subplot(3, 3, 5)
plt.scatter(y_test_pred, residuals, alpha=0.5, s=20)
plt.axhline(y=0, color='r', linestyle='--', linewidth=2)
plt.title('Residual Plot (Test Set)', fontsize=14, fontweight='bold')
plt.xlabel('Predicted Yield', fontsize=12)
plt.ylabel('Residuals', fontsize=12)
plt.grid(True, alpha=0.3)

# 6. Performance Comparison
ax6 = plt.subplot(3, 3, 6)
metrics_df = pd.DataFrame({
    'Train': [train_metrics['RMSE'], train_metrics['MAE'], train_metrics['R2']],
    'Validation': [val_metrics['RMSE'], val_metrics['MAE'], val_metrics['R2']],
    'Test': [test_metrics['RMSE'], test_metrics['MAE'], test_metrics['R2']]
}, index=['RMSE', 'MAE', 'R²'])

x = np.arange(len(metrics_df.index))
width = 0.25
plt.bar(x - width, metrics_df['Train'], width, label='Train', alpha=0.8)
plt.bar(x, metrics_df['Validation'], width, label='Validation', alpha=0.8)
plt.bar(x + width, metrics_df['Test'], width, label='Test', alpha=0.8)
plt.xlabel('Metrics', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.title('Performance Metrics Comparison', fontsize=14, fontweight='bold')
plt.xticks(x, metrics_df.index)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3, axis='y')

# 7. Prediction Error Distribution
ax7 = plt.subplot(3, 3, 7)
errors = np.abs(y_test - y_test_pred)
plt.hist(errors, bins=50, edgecolor='black', alpha=0.7, color='orange')
plt.title('Absolute Error Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Absolute Error', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.axvline(x=errors.mean(), color='r', linestyle='--',
            linewidth=2, label=f'Mean: {errors.mean():.2f}')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3, axis='y')

# 8. Sample Predictions Comparison
ax8 = plt.subplot(3, 3, 8)
sample_size = 50
indices = np.random.choice(len(y_test), sample_size, replace=False)
x_pos = np.arange(sample_size)
plt.plot(x_pos, y_test[indices], 'o-', label='Actual', linewidth=2, markersize=6)
plt.plot(x_pos, y_test_pred[indices], 's-', label='Predicted', linewidth=2, markersize=6)
plt.title(f'Sample Predictions (n={sample_size})', fontsize=14, fontweight='bold')
plt.xlabel('Sample Index', fontsize=12)
plt.ylabel('Yield', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

# 9. Q-Q Plot for Residuals
ax9 = plt.subplot(3, 3, 9)
from scipy import stats
stats.probplot(residuals, dist="norm", plot=plt)
plt.title('Q-Q Plot of Residuals', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('/content/model_performance.png', dpi=300, bbox_inches='tight')
print("Saved visualization: model_performance.png")

# ============================================================================
# 10. FEATURE IMPORTANCE ANALYSIS
# ============================================================================
print("\n[10] Feature Importance Analysis...")

# Create a simple permutation-based feature importance
feature_names = [
    'Crop_Year', 'Area', 'Annual_Rainfall', 'Fertilizer',
    'Pesticide', 'Crop', 'Season', 'State'
]

# Calculate baseline performance
baseline_mae = mean_absolute_error(y_test, y_test_pred)

importance_scores = []
for i in range(X_test_scaled.shape[1]):
    # Permute one feature
    X_test_permuted = X_test_scaled.copy()
    X_test_permuted[:, i] = np.random.permutation(X_test_permuted[:, i])
    X_test_permuted_cnn = X_test_permuted.reshape(X_test_permuted.shape[0],
                                                   X_test_permuted.shape[1], 1)

    # Predict with permuted feature
    y_pred_permuted = model.predict(X_test_permuted_cnn, verbose=0).flatten()
    permuted_mae = mean_absolute_error(y_test, y_pred_permuted)

    # Importance = increase in error
    importance = permuted_mae - baseline_mae
    importance_scores.append(importance)

# Visualize feature importance
plt.figure(figsize=(12, 6))
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importance_scores
}).sort_values('Importance', ascending=True)

plt.barh(importance_df['Feature'], importance_df['Importance'], color='steelblue')
plt.xlabel('Importance Score (Increase in MAE)', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.title('Feature Importance (Permutation-based)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, axis='x')
plt.tight_layout()
plt.savefig('/content/feature_importance.png', dpi=300, bbox_inches='tight')
print("Saved visualization: feature_importance.png")

# ============================================================================
# 11. SAVE MODEL AND RESULTS
# ============================================================================
print("\n[11] Saving Model and Results...")

# Save the model
model.save('/content/crop_yield_1d_cnn_model.h5')
print("Saved model: crop_yield_1d_cnn_model.h5")

# Save the scaler
import pickle
with open('/content/scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
print("Saved scaler: scaler.pkl")

# Save label encoders
with open('/content/label_encoders.pkl', 'wb') as f:
    pickle.dump(label_encoders, f)
print("Saved label encoders: label_encoders.pkl")

# Create results summary
results_summary = f"""
{'='*80}
CROP YIELD PREDICTION - 1D CNN MODEL RESULTS
{'='*80}

Dataset Information:
-------------------
- Total Samples: {len(df)}
- Training Samples: {len(X_train)}
- Validation Samples: {len(X_val)}
- Test Samples: {len(X_test)}
- Features: {len(feature_names)}

Features Used:
-------------
{', '.join(feature_names)}

Model Architecture:
------------------
- Type: 1D Convolutional Neural Network
- Convolutional Layers: 3
- Dense Layers: 4
- Total Parameters: {model.count_params():,}

Training Configuration:
----------------------
- Optimizer: Adam (lr=0.001)
- Loss Function: Huber Loss
- Batch Size: 64
- Max Epochs: 150
- Early Stopping: Yes (patience=20)
- Learning Rate Reduction: Yes (patience=10)

Performance Metrics:
-------------------
Training Set:
  - RMSE: {train_metrics['RMSE']:.4f}
  - MAE:  {train_metrics['MAE']:.4f}
  - R²:   {train_metrics['R2']:.4f}

Validation Set:
  - RMSE: {val_metrics['RMSE']:.4f}
  - MAE:  {val_metrics['MAE']:.4f}
  - R²:   {val_metrics['R2']:.4f}

Test Set:
  - RMSE: {test_metrics['RMSE']:.4f}
  - MAE:  {test_metrics['MAE']:.4f}
  - R²:   {test_metrics['R2']:.4f}

Top 3 Most Important Features:
------------------------------
{importance_df.tail(3).to_string(index=False)}

Model Files Saved:
-----------------
- crop_yield_1d_cnn_model.h5 (Trained model)
- scaler.pkl (Feature scaler)
- label_encoders.pkl (Categorical encoders)
- model_performance.png (Performance visualizations)
- feature_importance.png (Feature importance plot)

{'='*80}
"""

# Save results summary
with open('/content/model_results_summary.txt', 'w') as f:
    f.write(results_summary)

print(results_summary)

print("\n" + "="*80)
print("MODEL TRAINING COMPLETED SUCCESSFULLY!")
print("="*80)
print("\nAll files saved in /content/")
print("\nYou can use the saved model for predictions on new data.")
```
